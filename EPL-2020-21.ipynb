### Importing the necessary Python Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

### Importing the EPL 2020-21 dataset

```python
df= pd.read_csv('EPL_20_21.csv')
df.head(5)
```

### Getting necessary information about our dataset <br>
1. shape: gives us the total number of rows and columns present in the dataset <br>
2. info(): provides basic information like column names, non null value count, datatypes, memory usage etc. <br>
3. describe(): gives descriptive statistical data like count, mean, std, min, max <br>
4. isna(): total null value count

```python
#Total Goals
total_goals = df['Goals'].sum()
print(total_goals)
```

```python
#Total Penalty
total_PenaltyGoals = df['Penalty_Goals'].sum()
print(total_PenaltyGoals)
```

```python
#Total Penalty Attempts
total_PenaltyAttempts = df['Penalty_Attempted'].sum()
print(total_PenaltyAttempts)
```

```python
#Pie chart for penalties missed vs scored

plt.figure(figsize=(13,6))
pl_not_scored = total_PenaltyAttempts - total_PenaltyGoals
data = [pl_not_scored, total_PenaltyGoals]
label = ['Penalty Missed', 'Penalties Scored']
color = sns.color_palette('Blues')
plt.pie(data, labels = label, colors = color, autopct = '%0.f%%', explode = [0,0.1])
plt.show()
```

```python
#Which countries have the most number of players playing in the EPL

nationality = df.groupby('Nationality').size().sort_values(ascending = False)
nationality.head(10).plot(kind = 'bar', figsize=(12,6), color = sns.color_palette('deep'), xlabel='Country', ylabel = 'Number of players')
```

```python
#Clubs with maximum players in their squad

df['Club'].value_counts().nlargest(5).plot(kind = 'bar', color = sns.color_palette('deep'), xlabel ='Clubs', ylabel='Number of players')
```

```python
#Players based on age group
under20 = df[df['Age'] <=20]
age20_25 = df[(df['Age'] >20) & (df['Age']<=25)]
age25_30 = df[(df['Age'] >25) & (df['Age']<=30)]
above30 = df[df['Age'] >30]
```

```python
x = np.array([under20['Name'].count(), age20_25['Name'].count(),age20_25['Name'].count(), above30['Name'].count(),])
label = ['<=20', '>20 & <=25', '>25 & <=30', '>30']
plt.pie(x, labels = label, autopct = '%0.1f%%', radius = 1.5, colors = sns.color_palette('Reds'),explode=[0,0,0,0.1])
plt.show()
```

```python
under20['Club'].value_counts().plot(kind='bar',figsize=(18,6), color = sns.color_palette('Paired'))
plt.title('Clubs with maximum players under the Age of 20 ')
```

```python
#Average age of Players in each club

a = sns.boxplot(x='Club', y='Age', data=df)
a.set_xlabel('Club',fontsize=20)
a.set_ylabel('Age',fontsize=20)
plt.xticks(rotation = 90)
plt.show()
```

```python
#total assists from each club

asists_by_clubs = pd.DataFrame(df.groupby('Club', as_index=False)['Assists'].sum())
sns.set_theme(style='whitegrid', color_codes=True)
ax = sns.barplot(x='Club', y='Assists', data=asists_by_clubs.sort_values(by='Assists'))
ax.set_xlabel('Club',fontsize=20)
ax.set_ylabel('Assists',fontsize=20)
plt.xticks(rotation = 90)
plt.rcParams['figure.figsize']=(12,6)
plt.title("Plot of Club VS Total Assists", fontsize=20)
plt.show()
```

```python
#Top 10 Assists of the season

top10_assists = df[['Name','Club','Assists','Matches']].nlargest(n=10, columns='Assists')
top10_assists
```

```python
#total goals from each club

goals_by_clubs = pd.DataFrame(df.groupby('Club', as_index=False)['Goals'].sum())
sns.set_theme(style='whitegrid', color_codes=True)
ax = sns.barplot(x='Club', y='Goals', data=goals_by_clubs.sort_values(by='Goals'),palette='rocket')
ax.set_xlabel('Club',fontsize=20)
ax.set_ylabel('Goals',fontsize=20)
plt.xticks(rotation = 90)
plt.rcParams['figure.figsize']=(12,6)
plt.title("Plot of Club VS Total Goals", fontsize=20)
plt.show()
```

```python
#Top 10 Goal Scorers of the season

top10_goals = df[['Name','Club','Goals','Matches']].nlargest(n=10, columns='Goals')
top10_goals
```

```python
#Pie-Chart --Goals with and without assists

plt.figure(figsize=(14,7))
assists = df['Assists'].sum()
data = [total_goals - assists, assists]
labels = ['Goals without assists', 'Goals with assists']
color = sns.color_palette('Blues_r')
plt.pie(data, labels = labels, colors = color, autopct = '%.0f%%',explode=[0,0.02])
plt.show()
```

```python
#Top 10 players with the most yellow cards

epl_yellow = df.sort_values(by='Yellow_Cards', ascending = False)[:10]
plt.figure(figsize=(12,6))
plt.title('Players with the most Yellow Cards')
c = sns.barplot(x=epl_yellow['Name'], y=epl_yellow['Yellow_Cards'], label = 'Players', color='yellow')
plt.ylabel('Number of yellow cards')
plt.xticks(rotation = 75)
plt.show()
```

```python
#Top 10 players with the most Red cards

epl_red = df.sort_values(by='Red_Cards', ascending = False)[:10]
plt.figure(figsize=(12,6))
plt.title('Players with the most Red Cards')
c = sns.barplot(x=epl_red['Name'], y=epl_red['Red_Cards'], label = 'Players', color='red')
plt.ylabel('Number of red cards')
plt.xticks(rotation = 75)
plt.show()
```
